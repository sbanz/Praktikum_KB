```python
#Menampilkan INFO
import pandas as pd
import numpy as np

from prettytable import PrettyTable

df = pd.read_csv("Basketball_Players_Stats.csv")

df.info()
```

POINT 1 & 2

```python
inf_dat = {
    "Jumah Record"                 : df.shape[0],
    "Nama Attribute"               : list(df.columns.values),
    "Jumlhah Attribute Tipe Objek" : list(df.select_dtypes(["object"]).columns),
    "Jumlah Attribute Tipe Angka"  : list(df.select_dtypes(["int64", "float64"]).columns),
    "Jumlah Attribute Null"        : list(df.loc[:, df.isnull().any()].columns),
}

keys = list(inf_dat.keys())

inf_datTable = PrettyTable()
inf_datTable.field_names = ["Info", "Atribute", "Jumlah"]

inf_datTable.add_row([keys[0], "-", inf_dat[keys[0]]])

for i in range(1, 5):
    inf_datTable.add_row([keys[i], ", \n".join(inf_dat[keys[i]]), len(inf_dat[keys[i]])])
    
print(inf_datTable)
```
