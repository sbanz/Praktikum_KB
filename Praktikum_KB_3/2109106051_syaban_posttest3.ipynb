```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Basketball_Players_Stats.csv")

df.head(10)
```

```python
from numpy import size


season= df.dropna(subset=['Season']).sort_values(by=['Season'])['Season'].unique()

globalrank = df.groupby(['Season'])['GP'].sum()

plt.figure(figsize = (30,15))
plt.plot(season, globalrank, linestyle='--', marker='o', color='red')
plt.grid()
plt.title('Total Permainan Para Pemain Per Season')
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Penjualan (juta)', fontsize=12)
plt.fill_between(season, globalrank, color='lightgreen')


plt.show()
```

### JUMLAH GRAFIK MENUNJUKKAN BERAPA BANYAK TOTAL GAME YANG DIMAINKAN SETIAP PEMAIN DALAM SATU MUSIM

```python
Stages = df.dropna(subset=['Stage']).sort_values(by=['Stage'])['Stage'].unique()

total = df.groupby(['Stage'])['GP'].count()

patches, texts, pcts = plt.pie(total, labels=Stages, autopct='%.1f%%', wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       textprops={'size': 'x-large'})


plt.title('Total Pemain Basker Berdasarkan Stage')
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```

### GRAFIK MENUNJUKKAN BERAPA PERSEN TOTAL PERMAINAN BERDASARKAN STAGE DARI SELURUH DATA

```python
sns.set_style("ticks")
plt.figure(figsize = (25,9))
sns.countplot(data = df, x = "Season", hue = "Stage", palette = "rocket")
plt.title('Perbandingan panggung bermain')
```
